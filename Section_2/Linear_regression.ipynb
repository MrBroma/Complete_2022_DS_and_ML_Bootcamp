```python
import pandas as pd

from sklearn.linear_model import LinearRegression

# Load the dataset (clean data)
df = pd.read_csv('cost_revenue_clean.csv')
```

```python
# description of the dataframe
df.describe()
```

```python
# define the data/predictors as the pre-set features names.
X = pd.DataFrame(df, columns=['production_budget_usd'])

# Put the target in another DataFrame
y = pd.DataFrame(df, columns=['worldwide_gross_usd'])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(14, 8))
plt.scatter(X, y, alpha=0.3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')

plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)

plt.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

Slope Coefficient:

```python
regression.coef_ # theta_1
```

```python
# Intercept
regression.intercept_
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(X, y, alpha=0.3)
plt.plot(X, regression.predict(X), color='red', linewidth=2)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```
